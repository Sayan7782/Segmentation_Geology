```python
import tensorflow as tf
from keras.layers import *
from keras.models import Model,Sequential
from keras import backend as K
```

```python
#FULLY CONNECTED NETWORK
model=Sequential()
model.add(SeparableConv2D(9,(3,3),padding="same",activation=None,input_shape=(48,48,3)))
model.add(BatchNormalization(axis=-1))
model.add(LeakyReLU(alpha=0.3))

model.add(MaxPooling2D(pool_size=(2,2),strides=None))
model.add(BatchNormalization(axis=-1))

model.add(SeparableConv2D(36,(3,3),padding="same",activation=None))
model.add(BatchNormalization(axis=-1))
model.add(LeakyReLU(alpha=0.3))
          

model.add(MaxPooling2D(pool_size=(2,2),strides=None))
model.add(BatchNormalization(axis=-1))
          
model.add(SeparableConv2D(144,(3,3),padding="same",activation=None))
model.add(LeakyReLU(alpha=0.3))
model.add(BatchNormalization(axis=-1))

model.add(MaxPooling2D(pool_size=(2,2),strides=None))
model.add(BatchNormalization(axis=-1))


model.add(SeparableConv2D(576,(3,3),padding='same',activation=None))
model.add(BatchNormalization(axis=-1))
model.add(LeakyReLU(alpha=0.3))

model.add(MaxPooling2D(pool_size=(2,2),strides=None))
model.add(BatchNormalization(axis=-1))

model.add(SeparableConv2D(2304,(3,3),padding='same',activation=None))
model.add(BatchNormalization(axis=-1))
model.add(LeakyReLU(alpha=0.3))



model.add(Conv2DTranspose(1,(3,3),activation=None,padding='same',strides=(16,16)))         
model.add(Flatten())
model.add(BatchNormalization(axis=-1))
```

```python
model.summary()
```

```python
model1.add(UpSampling2D(size=(3,3), data_format=None, interpolation='nearest'))
model1.add(BatchNormalization(axis=-1))
model1.add(SeparableConv2D(576,(3,3),padding='same',activation=None))
model1.add(BatchNormalization(axis=-1))
model1.add(LeakyReLU(alpha=0.3))

model1.add(UpSampling2D(size=(2,2), data_format=None, interpolation='nearest'))
model1.add(BatchNormalization(axis=-1))
model1.add(SeparableConv2D(144,(3,3),padding='same',activation=None))
model1.add(BatchNormalization(axis=-1))
model1.add(LeakyReLU(alpha=0.3))

model1.add(UpSampling2D(size=(2,2), data_format=None, interpolation='nearest'))
model1.add(BatchNormalization(axis=-1))
model1.add(SeparableConv2D(36,(3,3),padding='same',activation=None))
model1.add(BatchNormalization(axis=-1))
model1.add(LeakyReLU(alpha=0.3))

model1.add(UpSampling2D(size=(2,2), data_format=None, interpolation='nearest'))
model1.add(BatchNormalization(axis=-1))
model1.add(SeparableConv2D(9,(3,3),padding='same',activation=None))
model1.add(BatchNormalization(axis=-1))
model1.add(LeakyReLU(alpha=0.3))

model1.add(UpSampling2D(size=(2,2), data_format=None, interpolation='nearest'))
model1.add(BatchNormalization(axis=-1))
model1.add(SeparableConv2D(1,(3,3),padding='same',activation=None))
model1.add(BatchNormalization(axis=-1))
model1.add(LeakyReLU(alpha=0.3))
model1.add((Flatten()))
```
