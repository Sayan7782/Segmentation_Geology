```python
import tensorflow as tf
from keras.layers import *
from keras.models import Model,Sequential
from keras import backend as K
```

```python
input=Input(shape=(48,48,3))
x=SeparableConv2D(9,(3,3),padding="same",activation=None,input_shape=(48,48,3))(input)
x1=BatchNormalization(axis=-1)(x)
x2=LeakyReLU(alpha=0.3)(x1)
x3=MaxPooling2D(pool_size=(2,2),strides=None)(x2)
x4=BatchNormalization(axis=-1)(x3)
x5=SeparableConv2D(36,(3,3),padding="same",activation=None)(x4)
x6=BatchNormalization(axis=-1)(x5)
x7=LeakyReLU(alpha=0.3)(x6)
x8=MaxPooling2D(pool_size=(2,2),strides=None)(x7)
x9=BatchNormalization(axis=-1)(x8)
x10=SeparableConv2D(144,(3,3),padding="same",activation=None)(x9)
x11=BatchNormalization(axis=-1)(x10)
x12=LeakyReLU(alpha=0.3)(x11)
x13=MaxPooling2D(pool_size=(2,2),strides=None)(x12)
x14=BatchNormalization(axis=-1)(x13)
x15=SeparableConv2D(576,(3,3),padding="same",activation=None)(x14)
x16=BatchNormalization(axis=-1)(x15)
x17=LeakyReLU(alpha=0.3)(x16)
x18=MaxPooling2D(pool_size=(2,2),strides=None)(x17)
x19=BatchNormalization(axis=-1)(x18)
x20=SeparableConv2D(2304,(3,3),padding="same",activation=None)(x19)
x21=BatchNormalization(axis=-1)(x20)
x22=LeakyReLU(alpha=0.3)(x21)
x23=MaxPooling2D(pool_size=(3,3),strides=None)(x22)
x24=BatchNormalization(axis=-1)(x23)
```

```python
x25=UpSampling2D(size=(3,3), data_format=None, interpolation='nearest')(x24)
x26=BatchNormalization(axis=-1)(x25)

x27=Add()([x26,x22])
x28=SeparableConv2D(576,(3,3),padding="same",activation=None)(x27)
x29=BatchNormalization(axis=-1)(x28)
x30=LeakyReLU(alpha=0.3)(x29)

x31=UpSampling2D(size=(2,2), data_format=None, interpolation='nearest')(x30)
x32=BatchNormalization(axis=-1)(x31)
x33=Add()([x32,x17])

x34=SeparableConv2D(144,(3,3),padding="same",activation=None)(x33)
x35=BatchNormalization(axis=-1)(x34)
x36=LeakyReLU(alpha=0.3)(x35)

x37=UpSampling2D(size=(2,2), data_format=None, interpolation='nearest')(x36)
x38=BatchNormalization(axis=-1)(x37)
x39=Add()([x38,x12])

x40=SeparableConv2D(36,(3,3),padding="same",activation=None)(x39)
x41=BatchNormalization(axis=-1)(x40)
x42=LeakyReLU(alpha=0.3)(x41)

x43=UpSampling2D(size=(2,2), data_format=None, interpolation='nearest')(x42)
x44=BatchNormalization(axis=-1)(x43)
x45=Add()([x44,x7])

x46=SeparableConv2D(9,(3,3),padding="same",activation=None)(x45)
x47=BatchNormalization(axis=-1)(x46)
x48=LeakyReLU(alpha=0.3)(x47)

x49=UpSampling2D(size=(2,2), data_format=None, interpolation='nearest')(x48)
x50=BatchNormalization(axis=-1)(x49)
x51=Add()([x50,x2])

x52=SeparableConv2D(1,(3,3),padding="same",activation=None)(x51)
x53=BatchNormalization(axis=-1)(x52)
x54=LeakyReLU(alpha=0.3)(x53)
x55=Flatten()(x54)
```

```python
model = Model(inputs=input, outputs=x55)
```
